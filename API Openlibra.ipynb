```python
import pandas as pd
from urllib.request import urlopen
import json

#EXTRACCIÓN DE LIBROS DEL AUTOR ADRIAN PAENZA CON EL API DE OPEN LIBRA (api pública)
url='https://www.etnassoft.com/api/v1/get/?book_author=Paenza&callback=?'
uclient=urlopen(url)
responsehtml=uclient.read()
uclient.close()
info=json.loads(responsehtml.decode()) #la respuesta se da en una lista de diccionarios, cada una para cada libro
df=pd.DataFrame(info) #creando un df para la info de todos los libros
print(list(df.columns)) #viendo las cabeceras de las columnas para filtrar solo algunas
df_paenza=df[['ID','title','content_short','publisher','publisher_date','pages','language']]
df_paenza
```
